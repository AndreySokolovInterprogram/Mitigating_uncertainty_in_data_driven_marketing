```python
# Serialization deserialization
import requests
import os
import pickle

record_id = '17048695'
filename_to_download = 'data.pkl'

api_url = f'https://zenodo.org/api/records/{record_id}'

try:
    print(f"Searching for file '{filename_to_download}' in Zenodo repository with ID: {record_id}...")

    response = requests.get(api_url)
    response.raise_for_status()
    data = response.json()

    file_url = None
    for file in data.get('files', []):
        if file.get('key') == filename_to_download:
            file_url = file.get('links', {}).get('self')
            break

    if file_url:
        print(f"File '{filename_to_download}' found. Starting download...")

        file_response = requests.get(file_url, stream=True)
        file_response.raise_for_status()

        # Save the file to the current directory
        with open(filename_to_download, 'wb') as f:
            for chunk in file_response.iter_content(chunk_size=8192):
                f.write(chunk)

        print(f"File '{filename_to_download}' downloaded successfully.")
    else:
        print(f"File '{filename_to_download}' not found in the repository.")

except requests.exceptions.RequestException as e:
    print(f"An error occurred while accessing Zenodo: {e}")


with open('data.pkl', 'rb') as file:
  uss = pickle.load(file)
```

```python
#We get a dictionary. The key is the ID of the group, the value is the list of users of this group.
from time import sleep
groups = [17700656, 99946133, 142786786, 89373035, 171840666]
usersOfAllGroups = dict()
for grp in groups:
  group_users = []
  offset = 0

  response = vk.groups.getMembers(group_id = grp, offset=0)
  len_1 = response["count"]
  # len_1 = 3000

  while offset < len_1:
    response = vk.groups.getMembers(group_id = grp, offset=offset)
    group_users += response["items"]
    offset += 1000
    sleep(1)
  usersOfAllGroups[grp] = group_users
```

```python
# We identify users with intersections of delivery groups (no with five, 4 with four users)
listOfsets = []
request = []
num = 4
cnt = 1
for i in s_UOAG.keys():
  if cnt <= num:
    request.append(i)
  cnt += 1
for grp in request:
  listOfsets.append(set(s_UOAG[grp]))

inter = set.intersection(*listOfsets)
inter
```
